```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io as sio
from sklearn.svm import SVC
```

```python
data = sio.loadmat('C:\\Users\\Administrator\\Desktop\\ex6data1.mat')
data
```

```python
x,y = data['X'],data['y']
x.shape,y.shape
```

```python
def plot_data(x,y):
    plt.scatter(x[:,0],x[:,1],c=y.flatten(),cmap='jet')
    plt.xlabel('x1')
    plt.ylabel('y1')
plot_data(x,y)
```

```python
svc1 = SVC(C=1,kernel='linear')
svc1.fit(x,y.flatten())
svc1.predict(x)
svc1.score(x,y.flatten())
```

```python
def plot_boundary(model):
    x_min,x_max = -0.5,4.5
    y_min,y_max = 1.3,5
    xx,yy = np.meshgrid(np.linspace(x_min,x_max,500),np.linspace(y_min,y_max,500))
    z = model.predict(np.c_[xx.flatten(),yy.flatten()])
    zz = z.reshape(xx.shape)
    plt.contour(xx,yy,zz)
```

```python
plot_boundary(svc1)
plot_data(x,y)
```

```python
svc100 = SVC(C=100,kernel='linear')
svc100.fit(x,y.flatten())
svc100.predict(x)
svc100.score(x,y.flatten())
```
